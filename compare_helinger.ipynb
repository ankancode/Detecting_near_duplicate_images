```python
import cv2
import numpy as np
import os
import shutil 
import multiprocessing as mp
import time
```

```python
directory = "/home/ankan/Projects/infilect/images"
hist_dic = {}
paths = []
hist_val = []
image_names = []
resized_images = "/home/ankan/Projects/infilect/Resized_224_224_no_duplicates"

if not os.path.exists(resized_images):
    os.mkdir(resized_images)
else:
    shutil.rmtree(resized_images)
    os.mkdir(resized_images)

    
for filename in os.listdir(directory):
    
    image_path =  os.path.join(directory,filename)
    image = cv2.imread(image_path)
    resize_image = cv2.resize(image,(224,224)) 
    resized_image_path = os.path.join(resized_images,filename)
    cv2.imwrite(resized_image_path, resize_image)
    
    
for filename in os.listdir(resized_images):
    image_path =  os.path.join(resized_images,filename)
    image = cv2.imread(image_path)

    image_hist = cv2.calcHist([image], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
    image_hist = cv2.normalize(image_hist, image_hist).flatten()
    hist_dic[image_path] = image_hist
    paths.append(image_path)
    hist_val.append(image_hist)
    image_names.append(filename)
    
hist_nd = np.array(hist_val)
```
